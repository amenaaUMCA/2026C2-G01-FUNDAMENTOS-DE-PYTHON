# Semana 07 - Procesamiento de listas y estructuras de datos

## Recurso inicial: lista, tupla, diccionario o set

| Estructura | Se escribe | Cuando usarla | Ejemplo en EmprendeCR | Cuidado principal |
|---|---|---|---|---|
| Lista | `[]` | Varios datos relacionados que se recorren o conservan orden. | Ventas: `[85000, 92000, 78000]` | No depender de indices sin explicar su significado. |
| Tupla | `()` | Par o grupo fijo que no se modifica. | `("Soda San Pedro", 462000)` | No tratarla como lista mutable. |
| Diccionario | `{}` | Registro con campos con nombre. | `{"nombre": "Soda San Pedro", "ventas": [...]}` | Usar claves exactas. |
| Set | `set()` | Valores unicos, sin repetidos. | Provincias analizadas. | No usarlo si el orden importa. |

Regla rapida: lista para recorrer, diccionario para describir, tupla para pares fijos y set para quitar repetidos.


## Caso resuelto: analisis de emprendimientos costarricenses

Este notebook refleja la solucion esperada del archivo `analisis_emprendimientos_solucion.py`. Se conserva como material de apoyo para repasar la secuencia construida en clase.

## Solucion completa

La solucion usa funciones para calcular, un ciclo para procesar todas las sedes, diccionarios para el reporte, un set para provincias, tuplas para ranking y una lista para conservar empates en la sede con mayor ingreso.

```python
"""Practica Semana 07: analisis de emprendimientos costarricenses.

Complete los espacios marcados con TODO. El objetivo es generar un reporte por
sede usando listas, diccionarios, funciones, ciclos y condicionales.
"""

# Se importa la lista de sedes desde el archivo sedes.py
from sedes import sedes


def calcular_total(ventas):
    """Recibo una lista, la sumo y retorno el total."""

    # sum() permite sumar todos los valores numéricos de una lista
    return sum(ventas)


def calcular_promedio(lista):
    """Retorna el promedio de ventas de una lista."""

    # El promedio se calcula dividiendo la suma total entre la cantidad de datos
    return sum(lista) / len(lista)


def calcular_porcentaje(total, meta):
    """Calcula el porcentaje de cumplimiento de una meta."""

    # Fórmula: total alcanzado dividido entre la meta, multiplicado por 100
    return total / meta * 100


def calcular_clasificacion(total, meta):
    """Clasifica la sede según el porcentaje de cumplimiento de la meta."""

    # Primero se calcula el porcentaje de cumplimiento
    porcentaje = calcular_porcentaje(total, meta)

    # Si el porcentaje es igual o mayor a 100, la sede cumplió la meta
    if porcentaje >= 100:
        mensaje_sede = "Meta alcanzada."

    # Si el porcentaje es igual o mayor a 80, pero menor que 100,
    # la sede está cerca de cumplir la meta
    elif porcentaje >= 80:
        mensaje_sede = "Meta casi alcanzada, prestar atención."

    # Si el porcentaje es menor a 80, la sede no alcanzó la meta
    else:
        mensaje_sede = "Meta no alcanzada URGE ATENCION."

    # Se retorna el mensaje final de clasificación
    return mensaje_sede


def imprimir_reporte(datos_reporte):
    """Imprime el reporte final de ventas por sede."""

    # Encabezado del reporte
    print("\nREPORTE FINAL")
    print("-" * 60)

    # Se recorre cada fila del reporte.
    # Cada fila es un diccionario con la información calculada de una sede.
    for fila in datos_reporte:
        print(f"Sede: {fila['nombre']}")
        print(f"Provincia: {fila['provincia']}")
        print(f"Tipo: {fila['tipo']}")

        # Se imprime el total con formato de moneda y sin decimales
        print(f"Total semanal: ₡{fila['total']:,.0f}")

        # Se imprime el promedio diario con formato de moneda y sin decimales
        print(f"Promedio diario: ₡{fila['promedio']:,.0f}")

        # Se imprime el porcentaje con dos decimales
        print(f"Cumplimiento: {fila['porcentaje']:.2f}%")

        print(f"Estado: {fila['estado']}")
        print("-" * 60)

    # Se muestra la cantidad total de sedes procesadas
    print("Cantidad de sedes:", len(datos_reporte))


# Lista donde se guardará el reporte final de cada sede
reporte = []

# Set donde se guardarán las provincias sin repetir
provincias = set()

# Lista donde se guardará una tupla con el nombre de la sede y su total de ventas
ranking = []

# Variable para guardar el monto de venta más alto encontrado
venta_mas_alta = 0

# Lista para guardar el nombre de la sede o sedes con más ingresos
sedes_mas_ingresos = []


# Se recorre la lista de sedes importada desde sedes.py
for sede in sedes:
    # Se extraen las ventas y la meta de la sede actual
    ventas = sede["ventas"]
    meta = sede["meta"]

    # Se calcula el total de ventas de la sede
    total_sede = calcular_total(ventas)

    # Se calcula el promedio diario de ventas
    promedio_sede = calcular_promedio(ventas)

    # Se calcula el porcentaje de cumplimiento de la meta
    porcentaje_sede = calcular_porcentaje(total_sede, meta)

    # Se obtiene la clasificación de la sede según su porcentaje
    estado = calcular_clasificacion(total_sede, meta)

    # Se agrega al reporte un diccionario con la información procesada
    reporte.append(
        {
            "nombre": sede["nombre"],
            "provincia": sede["provincia"],
            "tipo": sede["tipo"],
            "total": total_sede,
            "promedio": promedio_sede,
            "porcentaje": porcentaje_sede,
            "estado": estado,
        }
    )

    # Se agrega la provincia al set.
    # Si la provincia ya existe, no se repite.
    provincias.add(sede["provincia"])

    # Se agrega una tupla al ranking con el nombre de la sede y el total vendido
    ranking.append((sede["nombre"], total_sede))

    # Si el total de la sede actual es mayor que la venta más alta registrada,
    # se actualiza la venta más alta y se reinicia la lista de sedes ganadoras.
    if total_sede > venta_mas_alta:
        venta_mas_alta = total_sede
        sedes_mas_ingresos = [sede["nombre"]]

    # Si el total de la sede actual es igual a la venta más alta,
    # significa que hay empate, por eso se agrega también a la lista.
    elif total_sede == venta_mas_alta:
        sedes_mas_ingresos.append(sede["nombre"])


# Se imprime el reporte detallado de todas las sedes
imprimir_reporte(reporte)


# Se imprime el resumen general
print("\nRESUMEN FINAL")

# Muestra las provincias analizadas sin repetir
print("Provincias analizadas:", provincias)

# Muestra el ranking base con nombre de sede y total vendido
print("Ranking base:", ranking)

# Muestra el monto de venta más alto
print(f"Venta más alta: ₡{venta_mas_alta:,.0f}")

# Muestra la sede o sedes que tuvieron el ingreso más alto.
# El operador * desempaqueta la lista para imprimir sus elementos separados.
print(
    "Sedes con más ingresos:",
    *sedes_mas_ingresos,
)
```
